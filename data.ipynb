```python
import matplotlib.pyplot as plt

# 
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 

# ---  ---
hours = list(range(1, 25))
elec = [15, 25, 26, 28, 28, 32, 35, 41, 43, 49, 55, 68, 66, 56, 43, 42, 47, 51, 53, 54, 50, 39, 30, 27]
heat = [45, 24, 38, 48, 51, 58, 43, 58, 68, 83, 80, 84, 70, 68, 65, 62, 65, 88, 50, 48, 45, 55, 63, 50]
gas = [10, 13, 12, 11, 14, 15, 14, 15, 13, 10, 12, 15, 15, 15, 14, 13, 11, 11, 12, 12, 13, 10, 9, 8]
h2 = [80, 66, 60, 50, 50, 55, 55, 100, 152, 136, 80, 70, 70, 70, 90, 90, 80, 70, 70, 70, 70, 70, 70, 70]

# ---  ---
#  609x521 px ( =  / DPI)
fig, ax1 = plt.subplots(figsize=(6.09, 5.21), dpi=100)

#  ()
p1, = ax1.plot(hours, elec, 's-', color='#555555', label='', markersize=4, linewidth=1)
p2, = ax1.plot(hours, heat, '*-', color='#E31A1C', label='', markersize=5, linewidth=1)
p3, = ax1.plot(hours, gas, '^-', color='#F781BF', label='', markersize=4, linewidth=1)

ax1.set_xlabel(' (hour)', fontsize=10)
ax1.set_ylabel(' (MW)', fontsize=10)
ax1.set_ylim(0, 100)
ax1.tick_params(axis='both', direction='in', top=True)

#  ()
ax2 = ax1.twinx()
p4, = ax2.plot(hours, h2, 's-', color='#00BFFF', label='', markersize=4, linewidth=1)

ax2.set_ylabel(' (kg)', color='#00BFFF', fontsize=10)
ax2.set_ylim(40, 160)
ax2.tick_params(axis='y', colors='#00BFFF', direction='in')

# 
lines = [p1, p2, p3, p4]
ax1.legend(lines, [l.get_label() for l in lines], loc='upper left', fontsize=9, frameon=False)

plt.title('24', fontsize=12)
plt.tight_layout()
plt.show()
```
